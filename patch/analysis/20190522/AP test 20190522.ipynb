```python
def plot_ap(ap_file):
    abf = pyabf.ABF(ap_file)
    fig = pyabf.plot.sweeps(abf, title = False, offsetXsec=.15, offsetYunits=30)
    pyabf.plot.scalebar(abf)
    return fig
```

```python
apFiles = []
for apDir in ap_dirs:
    for ap_dirpath, _, ap_file_here in os.walk(os.path.join('AP',apDir)):
        for ap_file in ap_file_here:
            apFileFull = os.path.join(ap_dirpath, ap_file)
            apFiles.append(apFileFull)
```

Analyse AP count, amplitude and frequence:

```python
def get_ap_count(ap_file):
    abf = pyabf.ABF(ap_file)
    counts = []
    for sweep in abf.sweepList:
        abf.setSweep(sweep)
        counts.append(len(ap.ap_points_currentSweep(abf)))
    return counts
get_ap_count(apFiles[0])
```

```python
import efel
```

```python
def to_efelData(ap_file):
    abf = pyabf.ABF(ap_file)
    traces = []
    stim_start = abf.sweepEpochs.p1s[2]
    stim_end = abf.sweepEpochs.p2s[2]
    for sweep in abf.sweepList:
        abf.setSweep(sweep)
        trace = {}
        trace['T'] = abf.sweepX
        trace['V'] = abf.sweepY
        trace['stim_start'] = [stim_start]
        trace['stim_end'] = [stim_end]
        traces.append(trace)
    return(traces)
```
